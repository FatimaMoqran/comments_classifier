```python
#Importing libraries

import pickle
import csv
import pandas as pd
import nltk
from spellchecker import SpellChecker
from nltk.stem import WordNetLemmatizer
from french_lefff_lemmatizer.french_lefff_lemmatizer import FrenchLefffLemmatizer
```

```python
#Loading the picke file to a variable
vocab = pickle.load(open("itosref30k.pkl", "rb"))

#How long is the vocabulary ?
len(vocab)
```

```python
#Function to check word existence in dictionnary downloaded from https://drive.google.com/drive/folders/1_0D3zv5H7iMW1qk7wHNN1yQrHi6arv8g

#Initializing a list to store unknown words
unknown_words = []

#Writting the function
def wordChecker(word):
        
    if word in vocab:
        #no action required
        pass
    
    elif word in unknown_words:
        #no action required
        pass

    else:
        #print(("{} is not in the dictionnary").format(word))
        #Adding it to a list
        unknown_words.append(word)
```

```python
#Function to lemmatize 

#Instantiate lemmatizer
lemmatizer = FrenchLefffLemmatizer()

def word_lemmatizer(text):
       
    #lem_text = [lemmatizer.lemmatize(i) for i in text]
    
    return lemmatizer.lemmatize(text)
    
```

```python
#Dropping first columns (ids, labels) to keep comments only
df = pd.read_csv("df_csv.csv", header = None, encoding='utf-8')

#Keep only comments column
df_comments = df[3]

#Casting pandas object to string dtype
comments = df[3].astype(str)

#Converting the Serie to a whole string
cmts = '\n'.join(comments)
```

```python
#Tokenization

#Building tokenizer
tokenizer = nltk.RegexpTokenizer(r'\w+')

#Tokenizing each comment
tokens = comments.apply(lambda x: tokenizer.tokenize(x.lower()))
tokens
```

```python
#Lemmatization
coments_lemmatized = comments.apply(lambda x : word_lemmatizer(x))
coments_lemmatized
```

```python
#Words checking

#Building method for French language
spell = SpellChecker(language='fr')

#Nested loop to check every word in each comment
for i in range(len(tokens)):

    #Applying the function to check if word in vocab dictionnary
    for word in tokens[i]:
        wordChecker(word)

    #Get the one 'most likely' answer
    #print(spell.correction(word))

    #Get a list of 'likely' options
    #print(spell.candidates(word))
```
